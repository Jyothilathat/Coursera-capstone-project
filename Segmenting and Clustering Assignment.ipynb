Step 1: Importing the required libraries

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
import lxml
```

Step 2: Downloading the required data using the provided link

```python
#converting url into html
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
Datasource=requests.get(url).text
#converting html into xml
soup = BeautifulSoup(Datasource,'lxml')
table=soup.find('table')    
```

Step 3: Creating pandas dataframe

```python
#Assigning rows and storing its data
tablerows=table.find_all('tr')
tabledata=[]
for row in tablerows:
    td=[]
    for cell in row.find_all('td'):
        td.append(cell.text.strip())
        tabledata.append(td)
        
#Creating pandas dataframe by removing not assigned, resetting the index, grouping, resetting the index       
df_table=pd.DataFrame(tabledata, columns=['PostalCode','Borough','Neighborhood'])
df_table=df_table[~df_table['Borough'].isnull()]
df_table.drop(df_table[df_table.Borough == 'Not assigned'].index, inplace=True)#ignoring not assigned cells
df_table.reset_index(drop=True, inplace=True)
df_table=df_table.groupby(['PostalCode','Borough'])['Neighborhood'].apply(lambda x: ','.join(x)).reset_index()
df_table['Neighborhood'].replace('Not assigned',df_table['Borough'],inplace=True)
```
